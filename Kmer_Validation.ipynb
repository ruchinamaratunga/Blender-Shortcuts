```python
def kmer_counter(kmers, file_path, k, read_limit=None):
    fp = open(file_path, "r")
    
    if (read_limit != None):
        current_read = 0
        while (current_read < read_limit):
        
            current_read += 1
            # Get next line from file
            line = fp.readline()
            # print(len(line))
            # if line is empty
            # end of file is reached
            if not line:
                break

            pointer = 0
            while (pointer+k < len(line)):
                kmer = line[pointer:pointer+k]
                if kmer in kmers:
                    kmers[kmer] += 1
                else:
                    kmers[kmer] = 1

                pointer = pointer+1

    else: 
        while True:
            # Get next line from file
            line = fp.readline()
        
            # if line is empty
            # end of file is reached
            if not line:
                break

            pointer = 0
            while (pointer+k < len(line)):
                kmer = line[pointer:pointer+k]
                if kmer in kmers:
                    kmers[kmer] += 1
                else:
                    kmers[kmer] = 1

                pointer = pointer+1
```

```python
# takes kmer = a dictionary of kmer counts and output_file_path
# dump the dictionary into a file
def dump_kmers(kmers, output_file_path):
    f1 = open(output_file_path, "w")
    for k, v in kmers.items():
        f1.write(str(k)+" : "+str(v)+"\n")
```

```python
# take kmer output file path and return aggregation of counts
# this should be equal to individual counts
def count_total_counts(output_file):
    f1 = open(output_file, "r")
    l1 = f1.readlines()
    total = 0
    for i in l1:
        kmer, count = i.split(" : ")
        count = int(count)
        total += count

    return total
```

```python
def compare_kmer_results(kmer_dict, cpp_output_file):
    f1 = open(cpp_output_file, "r")

    l1 = f1.readlines()
    
    if(len(l1) != len(kmer_dict)):
        return "Unmatch unique kmer amount"

    for i in l1:
        kmer, count = i.split(" : ")
        count = int(count)
        if (kmer_dict.get(kmer) == None):
            return "Unmatched Kmers Exists"
        else:
            if (count != kmer_dict.get(kmer)):
                return "Unmatch Count exists"

    return True
```
